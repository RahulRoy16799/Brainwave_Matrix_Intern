# Exploratory Data Analysis

- <H4> WHAT IS THE OVERALL SALES TREND?

```python
#Minimum order date
df['order_date'].min()
```

```python
#Maximum order date
df['order_date'].max()
```

```python
# Convert 'order_date' to datetime format
df['order_date'] = pd.to_datetime(df['order_date'])

# Extract 'month_year' from the data set
df['month_year'] = df['order_date'].dt.strftime('%Y-%m')

# Display result
print(df)
```

```python
df.groupby('month_year').sum()
```

```python
#'sales' column is stored as an object (string) instead of a numeric type.So,converting the 'sales' column to a numeric format before groupby() operation
df['sales'] = pd.to_numeric(df['sales'], errors='coerce')
df.groupby('month_year').sum()['sales'] 
```

```python
#Grouping month year with sales
df_trend =df.groupby('month_year').sum()['sales'].reset_index()
print(df_trend)
```

```python
plt.figure(figsize=(15,6)) #Setting the figure size as the graph is small
plt.plot(df_trend['month_year'], df_trend['sales'], color='#b80045') # Red Color code'#b80045'
plt.xticks(rotation='vertical',size=8) #Setting the figure 90 degree for better understanding
plt.show()
```

- <H4> WHICH ARE THE TOP 10 PRODUCTS BY SALES?

```python
df.head(3)
```

```python
df['product_name']
```

```python
#Grouping the product_name column by sale.
prod_sales = pd.DataFrame(df.groupby('product_name').sum()['sales'])
```

```python
#Sorting prod_sales column
prod_sales = prod_sales.sort_values('sales',ascending=False)
```

```python
#Top 10 products by sales
prod_sales[:10]
```

- <H4> WHICH ARE THE MOST SELLING PRODUCTS?

```python
#Grouping product name
most_sell_prod = pd.DataFrame(df.groupby('product_name').sum()['quantity'])
```

```python
#Sorting most_sell_products
most_sell_prod = most_sell_prod.sort_values('quantity',ascending=False)
```

```python
#Top 10 most selling products 
most_sell_prod[:10]
```
